```python
import pandas as pd
from bs4 import BeautifulSoup
import os
import matplotlib.pyplot as plt
```

# reading the pages from folder 

```python
def read_files(folder_path, encoding='utf-8'):
    content_list = []
    
    for file_path in folder_path:
        try:
            with open(file_path, 'r', encoding=encoding) as file:
                html_content = file.read()
                content_list.append(html_content)
        except UnicodeDecodeError:
            print(f"Error decoding file: {file_path}")

    return content_list
```

# read model of the cars from html pages 

```python
def car_Model_Info(page):
    soup = BeautifulSoup(page, 'lxml')
    info=soup.find('table',{'class':'driving-table'})
    if info:
        infoModel=info.find_all('h5')
        if infoModel:
            element=infoModel[0].text.strip()[9:]
        return element
```

# read price of the cars from html pages 

```python
def car_price_Info(page):
    soup = BeautifulSoup(page,'lxml')
    price = soup.find('table',{'class':'driving-table'})

    if price :
        priceInfo=price.find_all('h5')[1].text.strip().split()[0]
        return priceInfo
```

# read name of the cars from html pages 

```python
def car_name_Info(page):
    soup = BeautifulSoup(page,'lxml')
    name = soup.find('table',{'class':'driving-table'})

    if name :
        nameInfo=name.find_all('h3')[0].text.strip()
        return nameInfo
```

# read the features of the cars from html pages without اضافات

```python
def car_feature_Info(page):
    soup = BeautifulSoup(page, 'lxml')
    all_features = soup.find('table', {'class': 'list_ads'})
    features_dict = {}

    if all_features:
        rows = all_features.find_all('tr')
        
        for row in rows:
            columns = row.find_all('td')
            
            if len(columns) == 2:
                key = columns[0].text.strip()
                value = columns[1].text.strip()
                features_dict[key] = value
                  
                if key == 'إضافات':
                    # Check if the key exists before attempting to remove it
                    if key in features_dict:
                        del features_dict[key]
                    

    return features_dict
```

# read اضافات of the cars from html pages 

```python
def car_extra_Info(page):
    soup = BeautifulSoup(page,'lxml')
    extra = soup.find('td',{'class':'list-additions'})

    if extra :
         extraInfo=extra.find_all('li')
         add=[li.text.strip() for li in extraInfo]
         return add
```

# read the isurance info

```python
def car_insurance(page):
    soup = BeautifulSoup(page, 'lxml')
    
    # Find all tables with the class 'list_ads'
    insurance_tables = soup.find_all('table', {'class': 'list_ads'})
    
    if insurance_tables:
        insurance_table = insurance_tables[1]
        insurance_dict = {}

        rows = insurance_table.find_all('tr')

        for row in rows:
            columns = row.find_all('td')

            if len(columns) == 2:
                key = columns[0].text.strip()
                value = columns[1].text.strip()
                insurance_dict[key] = value

        return insurance_dict

    return None
```

# calling the functions

```python
folder_path='D:/MLproject/data'
file_names = os.listdir(folder_path)
file_paths = [os.path.join(folder_path, file_name) for file_name in file_names]

pages = read_files(file_paths)
# for content in pages:
#     print(len(content))
```

```python
carsModel=[]
for page in pages :
    data= car_Model_Info(page)
    carsModel.append(data)
carsModel
```

```python
carsModel
```

```python
carsPrice=[]
for page in pages:
    price_data=car_price_Info(page)
    carsPrice.append(price_data)

carsPrice
```

```python
carsName=[]
for page in pages:
    name_data=car_name_Info(page)
    carsName.append(name_data)

carsName
```

```python
carFeatures = []

for page in pages:
    feature_data = car_feature_Info(page)
    carFeatures.append(feature_data)

carFeatures
```

```python
car_extraFeatures = []

for page in pages:
    extra_data = car_extra_Info(page)
    car_extraFeatures.append(extra_data)

car_extraFeatures
```

```python
car_insurance_info=[]
for page in pages:
    insurance_data = car_insurance(page)
    car_insurance_info.append(insurance_data)

car_insurance_info
```

# find the unique values of اضافات

```python
flat_list = [item for sublist in car_extraFeatures if sublist is not None for item in sublist]
unique_values = set(flat_list)
print(unique_values)
```

```python
binary_indicators_list = []

for sample_extra_features in car_extraFeatures:
    binary_indicators = {}
    if sample_extra_features is not None:
        for feature in sample_extra_features:
            binary_indicators[feature] = 1
    else:
        # If the list is None, set all indicators to 0
        binary_indicators = {}
    binary_indicators_list.append(binary_indicators)

print(binary_indicators_list)
```

# create DataFrame

```python
filtered_car_insurance_info = [info for info in car_insurance_info if info is not None]
filtered_carFeatures = [info for info in carFeatures if info is not None]

# Create DataFrames from the lists
car_insurance_df = pd.DataFrame(filtered_car_insurance_info)
car_features_df = pd.DataFrame(filtered_carFeatures)

# Merge DataFrames on a common column (you can adjust this based on your data)
cars_df = pd.concat([car_insurance_df, car_features_df], axis=1)

cars_df['name'] = carsName
cars_df['model'] = carsModel
cars_df['price'] = carsPrice

# Save the merged DataFrame to a CSV file
cars_df.to_csv('carsData.csv', index=False)

cars_df
```
